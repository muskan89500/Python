```python

# Check whether a number is prime or not
num = int(input("Enter a number: "))

if num <= 1:
    print("Not Prime Number")
else:
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print("Prime Number")
    else:
        print("Not Prime Number")

# Factorial Program
num = int(input("Enter a number: "))

fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```
