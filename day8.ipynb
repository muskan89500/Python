```python
#fibonacci series( the fibonacci series is the sequence of numbers in which each number is the sum of the two previous numbers)
```

```python
n = int(input("enter the numbers of terms: "))
a = 0
b = 1
print("fibonacci series: ")
for i in range(n):
    print(a, end=" ")
    c = a+b
    a = b
    b = c
```
